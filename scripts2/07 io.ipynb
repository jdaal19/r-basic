```python
x = int(input("Escribe un numero: "))
if x < 5:
    print("Has escrito un mumero menor que 5")
elif x < 10:
    print("Has escrito un mumero menor que 10 y mayor o igual a 5")
else:
    print("El numero que has escrito es mayor o igual que 10")
```

```python
x = int(input("Escribe un numero: "))
if x>0 and x <10:
    print("Has escrito un numero en el rango de [0,10]")
```
